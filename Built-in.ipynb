```python
import json

data = {"name": "علي", "city": "الرياض"}
json_text = json.dumps(data)

print(json_text)  # النتيجة: '{"name": "\u0639\u0644\u064a", "city": "\u0627\u0644\u0631\u064a\u0627\u0636"}'
```

```python
import statistics

grades = [85, 90, 78, 92, 88]
print(statistics.mean(grades))  # النتيجة: 86.6 (المتوسط)
```
