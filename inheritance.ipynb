```python
class A:
    def greet(self):
        print("مرحباً من الكلاس A")

class B:
    def greet(self):
        print("مرحباً من الكلاس B")

# هنا الترتيب مهم جداً: A أولاً ثم B
class C(A, B):
    pass

# إنشاء كائن من الكلاس C
obj = C()
obj.greet()

#دائما يتم استدعاء الكلاس الاول في الترتيب وهو ما يسمى (The Diamond problem)  
#هذا الترتيب يسمى MRO(Method Resolution Order)
```
