```python
import numpy as np

#قائمة الكلمات
text1 = "I Love Trello 2"
text2 = "Trello 2"

#قمنا بجعلها احرف صغيرة و تقسيمها
list1 = text1.lower().split()
list2 = text2.lower().split()

#طباعة القائمتين
print(f"list 1 :{list1}")
print(f"list 2 :{list2}")

bag_of_words = []
for word in (list1 + list2):
    if word not in bag_of_words:
        bag_of_words.append(word)
        
#قائمة للفيكتورين
v1 = []
v2 = []

#اللوب لاضافه 1 اذا كانت الكلمة موجوده و 0 اذا لم تكن موجوده في القائمتين
for word in bag_of_words:
    if word in list1:
        v1.append(1)
    else:
        v1.append(0)
        
    if word in list2:
        v2.append(1)
    else:
        v2.append(0)

#تحويل الفيكتور الى مصفوفه
v1 = np.array(v1)
v2 = np.array(v2)


print(f"vector 1 : {v1}")
print(f"vector 2 : {v2}")

#
dot_product = np.dot(v1, v2)
norm1 = np.linalg.norm(v1)
norm2 = np.linalg.norm(v2)

print(f"الضرب النقطي :{dot_product}")
print(f"الطول الخطي 1 :{norm1}")
print(f"الطول الخطي 2 :{norm2}")

cosine_similarity = dot_product / (norm1 * norm2)

print(f"Bag of words: {bag_of_words}")
print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Cosine Similarity: {cosine_similarity}")
```
